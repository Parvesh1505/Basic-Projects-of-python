# Guess The Number 

```python
# Now we are guessing a number which is generated by computer
from random import randint
def guess(x):
    name = str(input('Enter your Name Please:-'))
    print(f"Welcome {name} in Guess The Number Game.")
    random_number = randint(1,x)
    your_no = 0
    while your_no!= random_number:
        your_no = int(input(f'Guess your number in between 1 to {x}\n:-'))
        if your_no > random_number:
            print('your number is high with computer number.Lets try again')
        elif your_no < random_number:
          print('Your number is low with computer number.Lets Try again')
    
    print(f"Wow..congratulations {name} You have Guess the Right number {random_number}")
    
guess(20)          
```

# Cryptography

```python
def intro():
    print(".....Welcome to cryptography world Here we convert a word into secert language.....")
    msg = """....Please Select Method for encoding for your covience...
          1.caesar cipher
          2.vigenere cipher
          3. Exit"""       
    print(msg)
# now we have 2 method to encode or decode a word
def caesar_cipher():
    print('---------------------------------------')
    print("....This is Caesar cipher Method....")
    print('---------------------------------------')
    code = ''
    word = input('Please enter a word for encode:-')
    for i in word:
        if i.isalpha():
            value = ord(i)+4
            code = code + chr(value)
        else :
            code = code + i
    print('---------------------------------------')
    print(f'The encode of your word is {code}')
    print('---------------------------------------')
def vigenere_cipher():
    
    print("....This is Vigenere cipher Method....")
    code = ''
    word = input('Enter word for encode:-')
    key = input('Please Enter key for corresponding word:-')
    if len(key) != len(word):
        while len(key) != len(word):
            for i in key:
                key = key+i
                if len(key) == len(word):
                    break
        print('---------------------------------------')
        print(f'corresponding key for word by their according their length is {key}')
        print('---------------------------------------')
    # now we encode our word 
    for j,k in zip(word,key):
        if j.isalpha() and k.isalpha():
            if j.isupper() and k.isupper():
                value = (ord(j)+ord(k))%26 + 65
                code = code + chr(value)
            else:
                value = (ord(j)+ord(k))%26 + 97
                code = code + chr(value)
       
        elif j.isalpha() and k.isdigit():
            value = (ord(j)+ int(k))//10
            code = code + str(value)
        
        elif j.isdigit() and k.isalpha():
            value = (int(j)+ ord(k))//10
            code = code + str(value)
        
        else:
            value = (int(j)+int(k))//10
            code = code + str(value)
    print('---------------------------------------')
    print(f'The perfectly converted code of your word is {code}') 
    print('---------------------------------------')                
def main():
    intro()
    while True:
        num = input('press key for selecting a method')
        if num == '1':
            caesar_cipher()
        elif num == '2':
            vigenere_cipher()
        elif num == '3':
            print('---------------------------------------')
            print('This option exit the program')
            print('---------------------------------------')
            ans = input("enter 'y' for yes and 'n' for no ")
            if ans == 'y':
                print('Thankyou')
                break
            else:
                continue
        else:
            print('Error Please Select correct Option')      
                     
        
main()

```
